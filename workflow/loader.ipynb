```python
import logging

# Just get logger - don't call setup_logging again
logger = logging.getLogger(__name__)

# Base class for all loader implementations
class AbstractLoader:
    def __init__(self, transformed_data):
        # Store the transformed DataFrame to be written to the sink
        self.transformed_data = transformed_data

    def sink(self):
        # Abstract method - must be implemented by subclasses
        pass

# Loader for Customer Lifetime Value (CLV) data
class FirstLoader(AbstractLoader):
    def sink(self):
        try:
            logger.info("Loading CLV data to DBFS...")

            row_count = self.transformed_data.count()
            logger.info(f"  Total rows to load: {row_count}")

            # Write DataFrame to DBFS without partitioning
            get_sink_source(
            sink_type='dbfs',
            df=self.transformed_data,
            path="/Volumes/workspace/end_to_end_data_engineering_project/"
                 "end_to_end_data_engineering_output/customer_highest_revenue/",
            method="overwrite"
        ).load_data_frame()
            
            logger.info(f" Successfully loaded {row_count} rows to DBFS...")           
          
        except Exception as e:
            logger.error(f" Failed to load CLV data: {str(e)}", exc_info=True)
            raise
        

# Loader for Daily Revenue Trend data
class DailyRevenueTrendLoader(AbstractLoader):
    def sink(self):

        try:
            logger.info("Loading Daily Revenue Trend data to DBFS...")

            row_count = self.transformed_data.count()
            logger.info(f"  Total rows to load: {row_count}")

            # Configuration for partitioning by order_date
            para = {
            'partitionByColumns':['order_date']
            }
            # Use sink factory to get the appropriate DBFS writer
            # and persist the DataFrame in overwrite mode with partitioning
            get_sink_source(
                sink_type='dbfs_wpartition',
                df=self.transformed_data,
                path="/Volumes/workspace/end_to_end_data_engineering_project/"
                    "end_to_end_data_engineering_output/DailyRevenueTrend/",
                method="overwrite",
                para = para
            ).load_data_frame()
            
            # Additionally write to Delta table for query optimization
            get_sink_source(
                sink_type='delta',
                df=self.transformed_data,
                path="default.DailyRevenueTrend",
                method="overwrite"
            ).load_data_frame()

            logger.info(f" Successfully loaded {row_count} rows to DBFS...")

        except Exception as e:
            logger.error(f" Failed to load Daily Revenue Trend data: {str(e)}", exc_info=True)
            raise
        

# Loader for Top Cities By Revenue data
class TopCitiesByRevenueLoader(AbstractLoader):
    def sink(self):

        try:
            logger.info("Loading Top Cities By Revenue data to DBFS...")
            row_count = self.transformed_data.count()
            logger.info(f"  Total rows to load: {row_count}")

            # Configuration for partitioning by city (defined but not used in current sink_type)
            para = {
            'partitionByColumns':['city']
             }
            # Write DataFrame to DBFS partitioning
            get_sink_source(
                sink_type='dbfs',
                df=self.transformed_data,
                path="/Volumes/workspace/end_to_end_data_engineering_project/"
                    "end_to_end_data_engineering_output/TopCitiesByRevenue/",
                method="overwrite"
            ).load_data_frame()

            logger.info(f" Successfully loaded {row_count} rows to DBFS...")

        except Exception as e:
            logger.error(f" Failed to load Top Cities By Revenue data: {str(e)}", exc_info=True)
            raise

# Loader for How many customers are repeat buyers vs one-time buyers?
class repeatvsOneTimeCustomersLoader(AbstractLoader):
    def sink(self):
        try:
            logger.info("Loading repeatvsOneTimeCustomers data to DBFS...")
            row_count = self.transformed_data.count()
            logger.info(f"  Total rows to load: {row_count}")

            # Write DataFrame to DBFS without partitioning
            get_sink_source(
                sink_type='dbfs',
                df=self.transformed_data,
                path="/Volumes/workspace/end_to_end_data_engineering_project/"
                    "end_to_end_data_engineering_output/repeatvsOneTimeCustomers/",
                method="overwrite"
            ).load_data_frame()

            logger.info(f" Successfully loaded {row_count} rows to DBFS...")

        except Exception as e:
            logger.error(f" Failed to load repeatvsOneTimeCustomers: {str(e)}", exc_info=True)
            raise

# Loader for What percentage of orders are delivered vs cancelled?

class orderDeliveredVsCancelledLoader(AbstractLoader):

    def sink(self):
        try:
            logger.info("Loading repeatvsOneTimeCustomers data to DBFS...")
            row_count = self.transformed_data.count()
            logger.info(f"  Total rows to load: {row_count}")

            # Write DataFrame to DBFS without partitioning

            get_sink_source(
                sink_type= 'dbfs',
                df = self.transformed_data,
                path = '/Volumes/workspace/end_to_end_data_engineering_project/'
                       'end_to_end_data_engineering_output/orderDeliveredVsCancelled/',
                method="overwrite"
            ).load_data_frame()
            logger.info(f" Successfully loaded {row_count} rows to DBFS...")

        except Exception as e:
            logger.error(f" Failed to load repeatvsOneTimeCustomers: {str(e)}", exc_info=True)
            raise
# Which payment methods drive the most revenue?
class paymentMethodRevenueLoader(AbstractLoader):

    def sink(self):
        try:
            logger.info("Loading repeatvsOneTimeCustomers data to DBFS...")
            row_count = self.transformed_data.count()
            logger.info(f"  Total rows to load: {row_count}")

            # Write DataFrame to DBFS without partitioning

            get_sink_source(
                sink_type= 'dbfs',
                df = self.transformed_data,
                path = '/Volumes/workspace/end_to_end_data_engineering_project/'
                       'end_to_end_data_engineering_output/paymentMethodRevenue/',
                method = "overwrite"
            ).load_data_frame()
            logger.info(f" Successfully loaded {row_count} rows to DBFS")

        except Exception as e:
            logger.error(f" Failed to load paymentMethodRevenue: {str(e)}", exc_info=True)
            raise

# Loader 1: Revenue Summary
# Which product categories generate the most revenue?
class productCategoriesRevenueLoader(AbstractLoader):

    def sink(self):
        try:
            logger.info("Loading productCategoriesRevenue data to DBFS...")
            row_count = self.transformed_data.count()
            logger.info(f"  Total rows to load: {row_count}")

            get_sink_source(
                sink_type = 'dbfs',
                df = self.transformed_data,
                path = "/Volumes/workspace/end_to_end_data_engineering_project/end_to_end_data_engineering_output/productCategoriesRevenue/product_categories_revenue_summary",
                method = "overwrite"

            ).load_data_frame()
            logger.info("   Revenue summary loaded")

        except Exception as e:
            logger.error(f" Failed to load paymentMethodRevenue: {str(e)}", exc_info=True)
            raise

# Loader 2: Top 5 Products
class top5ProductsLoader(AbstractLoader):
    def sink(self):
        try:
            logger.info("Loading top5Products data to DBFS...")
            row_count = self.transformed_data.count()
            logger.info(f"  Total rows to load: {row_count}")
                        
            # Write DataFrame to DBFS without partitioning
            logger.info("  Loading top 5 products...")
            get_sink_source(
                sink_type = 'dbfs',
                df = self.transformed_data,
                path = "/Volumes/workspace/end_to_end_data_engineering_project/end_to_end_data_engineering_output/productCategoriesRevenue/top_5_products_by_category",
                method = "overwrite"

            ).load_data_frame()
            logger.info("   Top 5 products loaded")
            logger.info(f" Successfully loaded {row_count} rows to DBFS...")
        
        except Exception as e:
            logger.error(f" Failed to load paymentMethodRevenue: {str(e)}", exc_info=True)
            raise
```
